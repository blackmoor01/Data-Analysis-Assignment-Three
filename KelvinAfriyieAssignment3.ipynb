```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
gas_prices=pd.read_csv("gas_prices.csv")
```

## QUESTION ONE

```python
plt.figure(figsize=(15,20))
plt.plot(gas_prices["Year"],gas_prices["France"],marker="s",linestyle="--",label="France")
plt.plot(gas_prices["Year"],gas_prices["UK"],marker="*",linestyle="--", color="pink", label="UK")
plt.plot(gas_prices["Year"],gas_prices["Germany"],marker="s",linestyle="--", color="green", label="Germany")
plt.plot(gas_prices["Year"],gas_prices["South Korea"],marker="+",linestyle="--", color="black",label="South korea")
plt.xlabel("Year")
plt.ylabel("Countries")
plt.title("Gas Prices for various countries",fontsize=20)
plt.xticks(gas_prices["Year"], rotation=45);
plt.yticks(gas_prices["France"],rotation=45)
plt.grid()
plt.legend();
```

## QUESTION TWO

```python
Workers = [40, 35, 100, 70]
Education = ["Primary", "JHS", "SHS", "Tertiary"]
```

```python
plt.figure(figsize=(7,7))
plt.pie(Workers,labels=Education, colors=["Violet","Indigo","Yellow","Purple"],autopct="%.2f%%", explode=[0,0.08,0,0], 
        shadow = True, textprops={"fontsize":10, "color":"gray"} )

plt.legend()
plt.title("PIE PLOT",fontsize=16);
```

## QUESTION THREE

```python
xt = np.array([0,0.1,0.2,0.3,0.5,0.8,1])
yt = np.array([1.0,2.1,3.5,6.5,7.2,5.9,6.3])
zt = xt * yt
```

```python
plt.figure(figsize=(5,3))
plt.scatter(xt,yt,marker="s", label="yt")
plt.scatter(xt,zt, marker= "^", label="zt")
plt.xlabel("xt")
plt.ylabel("yt and zt")
plt.grid()
plt.title("SCATTER PLOT");
```
